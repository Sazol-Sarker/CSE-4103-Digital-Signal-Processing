```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#task 1
sig1=[1,3,5,9]#origin at 5

sig2=[2,3,9,5]#origin at 9

conv_reslt=np.convolve(sig1,sig2,mode="full")
print(conv_reslt)
plt.stem(conv_reslt)
plt.show()

orign1=2
orign2=2

row=4
col=4
arr=[]

for i in range(row):
  full_row=[]
  for j in range(col):
    full_row.append(sig1[i]*sig2[j])
  arr.append(full_row)
print(arr)

convo=[]
orign=-1
ith=0

for i in range(row):
  cur_rw=i
  ith+=1
  cur_col=0
 # print(ith,"ith")
  while cur_rw>=0:
      if cur_rw==orign2 and cur_col==orign1:
        orign=ith
        break
      cur_rw-=1
      cur_col+=1

for i in range(1,col,1):
  cur_rw=row-1
  ith+=1
  cur_col=i
#  print(ith,"ith")
  while cur_rw>=0 and cur_col<col:
      if cur_rw==orign2 and cur_col==orign1:
        orign=ith
        break
      cur_rw-=1
      cur_col+=1
      
print("Origin at : ",orign,"th = ",conv_reslt[orign-1])

n=col+row-1
t=np.arange(-orign+1,n-orign+1,1)
print("Convolved signal::")
plt.stem(t,conv_reslt)
plt.show()
```
