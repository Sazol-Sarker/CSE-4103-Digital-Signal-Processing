```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#task1
import math
t = np.linspace(0,10,100) 
#print(t)
#print(len(t))
A=1
f=40
theta=np.pi/2

Compx_sig =A *np.cos(theta)* np.exp(-1j* (2*np.pi * f * t + theta) )

#sig1= np.exp(-0.5*t )*np.cos(2*np.pi)
plt.figure(figsize = (15,10))
style.use('dark_background')
plt.plot(t,np.real(Compx_sig),'b-',linewidth = 3, label='real part')
plt.plot(t,np.imag(Compx_sig),'g-',linewidth = 3, label='imaginary part')
plt.xlim([-0.5,13])
plt.xlabel('Time',fontsize = 15)
plt.ylabel('Amplitude', fontsize = 15)
plt.title('Complex Signal',fontsize = 20)
plt.legend(fontsize = 15)
plt.show()
```

```python
import sympy as sp
from sympy import *
x=[1,2,3,1,2,2]
#print(x)
z=sp.symbols('z')
zt=sp.symbols('z_t')
z_t=x[0]

for k in range(1,len(x)):
    z_t=z_t+x[k]*z**-k

print("z-transformation : ",z_t)
n=len(x)-1
pole_eq=z**n
print(pole_eq)
poles=solve(pole_eq,z)
print("Poles: ",poles)

zero_eq=0
for i in range(len(x)):
  zero_eq=zero_eq+x[i]*z**n
  n=n-1
zeros=solveset(Eq(zero_eq,0))
print("Zeros: ",zeros)
zeros=zeros.evalf(5)

zero_val=[]
for i in zeros:
  i=str(i)
  i=i.replace('*I','j')
  i=i.replace(' ',"")
  i=complex(i)
  zero_val.append(i) 

np.imag(zero_val) 
plt.plot(np.real(zero_val) , np.imag(zero_val) , 'o')
plt.plot(0 , 0 , 'x')
plt.gca().add_patch(plt.Circle((0,0), radius=1, fill=False, alpha=.3))
```
