```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style 
```

**Generating and Plotting X1 signal**

```python
#Generating and Plotting X1 signal
x1 =np.array([1,2,2,1,1])
n1=np.array([0,1,2,3,4])
plt.figure(figsize=(20,10))
style.use('dark_background')
plt.subplot(2,2,1)
markerline,stemlines,baseline=plt.stem(n1,x1,label='X1')
plt.setp(stemlines)
plt.legend(fontsize=30)


#generating and plotting x2
x2 =np.array([3,2,1])
n2=np.array([0,1,2])
''' plt.figure(figsize=(20,10)) '''
style.use('dark_background')
plt.subplot(2,2,2)
markerline,stemlines,baseline=plt.stem(n2,x2,label='X2')
plt.setp(stemlines)
plt.legend(fontsize=30)


#Plotting X1 on positon 3
plt.subplot(2,2,3)
markerline,stemlines,baseline=plt.stem(n1,x1,label='X1')
plt.setp(stemlines)
plt.legend(fontsize=30)

#Flip/Fold x2 and plotting
f_x2=x2[::-1]
n3=np.array([-2,-1,0])
plt.subplot(2,2,4)
markerline,stemlines,baseline=plt.stem(n3,f_x2,label='Flipped X2')
plt.setp(stemlines)
plt.legend(fontsize=30)

plt.show()
```

```python
#applications of convolution
#-signal denoising...
```

```python
#Create signal
Srate=256 #hz
t=np.arange(0,3,1/50)
pnts=len(t)
''' print(pnts) '''
print(t)
```

```python
#Creating a noiseless signal
x=np.sin(2*np.pi * 2 *t)
''' x=np.sin(t) '''

#Creating a random noise
noise = 5*np.random.randn(pnts)

#Adding noise to signal
NoisySignal = x + noise
print(len(noise))
```

```python
#Plotting noisy signal[Pseudo-continous]

plt.figure(figsize=(20,12))
style.use('dark_background')
plt.plot(t,NoisySignal,label='Noisy signal')
plt.legend(fontsize=30)
plt.show()
```

```python
#Plotting noisy signal[Pseudo-continous]

plt.figure(figsize=(20,12))
style.use('dark_background')
plt.stem(t,NoisySignal,label='Noisy signal')
plt.legend(fontsize=30)
plt.show()
```

```python
#selecting filters for Denoising...
filter=2*np.ones(50)/10.0 # Moving average filter
print(filter)
```

```python
# filtering by convolation
filteredSignal=np.convolve(NoisySignal,filter,mode='same')
```

```python
#Plotting Filtered signal
plt.figure(figsize=(15,8))
plt.plot(filteredSignal,label='filtered signal')
plt.legend(fontsize=30)

plt.show()
```

```python
#Plotting Filtered signal
plt.figure(figsize=(30,15))
plt.stem(filteredSignal,label='filtered signal')
plt.legend(fontsize=30)

plt.show()
```
