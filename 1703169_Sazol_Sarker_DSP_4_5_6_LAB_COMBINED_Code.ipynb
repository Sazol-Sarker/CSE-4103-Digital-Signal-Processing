**Z Transformation**

```python
import sympy as sp
from sympy import *
x=[1,2,3,1,2,2]
#print(x)
z=sp.symbols('z')
zt=sp.symbols('z_t')
z_t=x[0]
for k in range(1,len(x)):
    z_t=z_t+x[k]*z**-k
print("z-transformation : ",z_t)
n=len(x)-1
pole_eq=z**n
print(pole_eq)
poles=solve(pole_eq,z)
print("Poles: ",poles)
zero_eq=0
for i in range(len(x)):
  zero_eq=zero_eq+x[i]*z**n
  n=n-1
zeros=solveset(Eq(zero_eq,0))
print("Zeros: ",zeros)
zeros=zeros.evalf(5)
zero_val=[]
for i in zeros:
  i=str(i)
  i=i.replace('*I','j')
  i=i.replace(' ',"")
  i=complex(i)
  zero_val.append(i) 
np.imag(zero_val) 
plt.plot(np.real(zero_val) , np.imag(zero_val) , 'o')
plt.plot(0 , 0 , 'x')
plt.gca().add_patch(plt.Circle((0,0), radius=1, fill=False, alpha=.3))
```

**Correcting amplitudes & converting indices to frequecies**

```python
#cycle 11th lab

#Generating the signal
srate=256 #hz
t=np.arange(0.,1.,1/srate)
amp=3
f=5 #hz
x=amp*np.sin(2*np.pi*f*t)#signal

#initialize fourier coefficient
X=np.zeros(len(x),dtype=complex)

for freq in range(0,len(t)):
  #create complex sin wave & compute dot product with signal
  csw=np.exp(-1j*2*np.pi*freq*t)
  X[freq]=np.sum(np.multiply(x,csw))

#extract amplitude
amps=2*np.abs(X)/len(t)

#in the above line of code we multiply by 2 to incorporate the negative frequencies
#of complex sinusoidal
```

```python
#converting indices to frequency
#there are 0 to 255 indices at which we generate the sine wave
#these are frequency indices
#not the frequency in hertz. Since we have n=256 indices including zeros
#theref0ore after (n/2)
#i.e 128 in
#in this case, we get the negative waveforms of the first 128 samples. It means after n=128,
#the frequencies are violating the nyquist criteria.
#thus in order to convert indices into hertz we need (n/2+1) or (n/2) linerarly spaced
#samples between zero &
#Nyquist (srate/2) as show below

import math
Nyquist=srate/2
Hz=np.linspace(0,Nyquist,math.floor(len(t)/2+1))
```

```python
#Plotting the signal & its fourier transform
plt.figure(figsize=(24,12)) # set the size of figure
plt.suptitle('Signal & its fourier transform',fontsize=40)
style.use('dark_background')

plt.subplot(2,2,1)
plt.plot(t,x,linewidth=3)
plt.title("Signal of amplitude ;3; and frequency '5 Hz'",fontsize=25)
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)

plt.subplot(2,2,2)
markerline,stemlines,baseline=plt.stem(Hz,amps[range(0,len(Hz))])
plt.setp(stemlines,'linewidth',3)
plt.xlim(0,10)
plt.ylim(0,5)
plt.title("Fourier transform of a signal",fontsize=25)
plt.xlabel('Frequency in Hz',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)

plt.show()
```

**plot signal & fourier transform**

```python
#task1
srate=256
t=np.arange(0.,1.,1/srate)
x1=3*np.sin(2*np.pi*2*t)
x2=5*np.sin(2*np.pi*4*t)
x3=7*np.sin(2*np.pi*6*t)
DC=5
x4=DC+x1+x2+x3

#plot signal & fourier transform

plt.figure(figsize=(24,14))
plt.suptitle('Signal & its fourier transform',fontsize=40)
style.use('dark_background')

plt.subplot(2,2,1)
plt.plot(t,x4,linewidth=3)
plt.title("Combined wave",fontsize=25)
plt.xlabel('time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)


#initialize fourier coefficient
X=np.zeros(len(x4),dtype=complex)

for freq in range(0,len(t)):
  #create complex sin wave & compute dot product with signal
  csw=np.exp(-1j*2*np.pi*freq*t)
  X[freq]=np.sum(np.multiply(x4,csw))

#extract amplitude
amps=2*np.abs(X)/len(t)

#in the above line of code we multiply by 2 to incorporate the negative frequencies
#of complex sinusoidal

Nyquist=srate/2
Hz=np.linspace(0,Nyquist,math.floor(len(t)/2+1))

#Indices VS Amplitude plot of FT
plt.subplot(2,2,2)
markerline,stemlines,baseline=plt.stem(amps)
plt.setp(stemlines,'linewidth',3)
plt.title("Fourier transform",fontsize=25)
plt.xlabel('Indices',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)

#energy density calculation =amps**2
energy=np.abs(amps)**2


#Energy density spactrum plot of FT
plt.subplot(2,2,3)
plt.plot(Hz,energy[range(0,len(Hz))])
plt.xlim(-1,15)
plt.ylim(0,110)
plt.title("Energy density spactrum of FT",fontsize=25)
plt.xlabel('Frequency',fontsize=20)
plt.ylabel('Energy Density',fontsize=20)

#Frequency VS Amplitude plot of FT
plt.subplot(2,2,4)
markerline,stemlines,baseline=plt.stem(Hz,amps[range(0,len(Hz))])
plt.setp(stemlines,'linewidth',3)
plt.xlim(0,15)
plt.ylim(0,10)
plt.title("Fourier transform",fontsize=25)
plt.xlabel('Frequency in Hz',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)
plt.show()
```

**Task 2-Ct Q2 solve**

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import style
```

```python
#taking signal
xn=[2,-2,1] #discrete periodic signal witih period N=3
N=3
c=[0,0,0] #coefficient store
#Method of finding fourier coefficients

for i in range(0,len(xn)):
  for n in range(0,len(c)):
    tmp=(1/N)*xn[n]
    tmp*=np.exp(-1j*2*np.pi*i*(n/N))
    c[i]+=tmp
  print("C",i," : ",c[i])
```

```python
#magnitude calculation
mag=[0,0,0]
#phase calculation
phase=[0,0,0]

for i in range(0,len(c)):
  tmp1=np.real(c[i])**2#real part
  tmp1+=np.imag(c[i])**2#imag part
  mag[i]=math.sqrt(tmp1)#magnitude
  tmp2=np.real(c[i])#x
  tmp3=np.imag(c[i])#y
  phase[i]=np.arctan(tmp3/tmp2)
  print("Magnitude of C",i," : ",mag[i])
  
print()
for i in range(0,len(phase)):
      print("phase of C",i," : ",phase[i]," Redian")
```

```python
#energy density for each coefficient
den=[0,0,0]
for i in range(0,len(mag)):
  den[i]=mag[i]**2
  #print("Energy density of C",i," : ",c[i]**2)# density |px|=|ck|**2

#plot signal
plt.figure(figsize=(24,12))
style.use('dark_background')
plt.subplot(2,2,1)
plt.plot([0,1,2],xn,linewidth=3)
plt.title('Signal',fontsize=25)
plt.xlabel('Time in sec',fontsize=20)
plt.ylabel('Amplitude',fontsize=20)

plt.subplot(2,2,2)
plt.plot([0,1,2],den,linewidth=3)
plt.title('Energy Density of Signal',fontsize=25)
plt.xlabel('Frequency',fontsize=20)
plt.ylabel('Energy density',fontsize=20)

plt.show()
```
